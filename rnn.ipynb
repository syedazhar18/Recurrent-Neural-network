```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
import warnings
warnings.filterwarnings("ignore")
```

```python
X = np.array([
    [10, 20, 30],
    [20, 30, 40],
    [30, 40, 50],
    [40, 50, 60]
])

y = np.array([40, 50, 60, 70])
```

```python
X = X.reshape((4, 3, 1))
```

```python
# --- STEP 2: Build the Simple RNN Model ---
print("--- Training Simple RNN ---")
rnn_model = Sequential()
# SimpleRNN layer with 50 'neurons'
rnn_model.add(SimpleRNN(50, activation='relu', input_shape=(3, 1)))
rnn_model.add(Dense(1)) # Output layer (predicts 1 number)
rnn_model.compile(optimizer='adam', loss='mse')
rnn_model.fit(X, y, epochs=200, verbose=0)
```

```python
# --- STEP 4: Predict ---
print("\n--- STEP 4: Predict ---")
print("Let's ask the model: What comes after 50, 60, 70? (Expected: 80)")

test_input = np.array([[50, 60, 70]])
test_input = test_input.reshape((1, 3, 1))  # reshape for RNN

```

```python
rnn_prediction = rnn_model.predict(test_input, verbose=0)

print("\nResults: Predicting what comes after 50, 60, 70 ---")
print("Predicted Answer:", rnn_prediction[0][0])
```

```python
print("--- Training LSTM ---")
lstm_model = Sequential()
lstm_model.add(LSTM(50, activation='relu', input_shape=(3, 1)))
lstm_model.add(Dense(1))

lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X, y, epochs=200, verbose=0)
```

```python
# --- STEP 3: Predict ---
test_input = np.array([[50, 60, 70]])
test_input = test_input.reshape((1, 3, 1))

lstm_prediction = lstm_model.predict(test_input, verbose=0)

```

```python
print("\n--- Results: Predicting what comes after 50, 60, 70 ---")
print("Correct Answer: 80")
print(f"LSTM Predicts: {lstm_prediction[0][0]:.2f}")
```
